```python
embeddings_dict = {}
with open('roberta_layer8_CC_ab_FEvocab.txt', newline='') as embeddings:
    embeddings_reader = csv.reader(embeddings, delimiter=' ')
    for row in embeddings_reader:
        embeddings_dict[row[0]] = row[1:]
```

```python
print(len(embeddings_dict["dejected"]))
```

```python
train_pair = []
train_score = []
with open('FE_pairs_human_rater_scores_nodupes_train.csv', newline='', encoding='utf-8-sig') as train:
    train_reader = csv.reader(train, delimiter=',')
    for row in train_reader:
        word_one, word_two, score = row[0], row[1], row[2]
        print(word_one, word_two, score)
        train_pair.append([embeddings_dict[word_one], embeddings_dict[word_two]])
        train_score.append(score)
```
